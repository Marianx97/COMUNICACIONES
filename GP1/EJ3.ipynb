## a)

Se puede expresar la densidad de potencia como:

$S = \frac{P_t}{4.\pi.d^2}$

Y la potencia recibida como:

$P_r = S . A_e$

Donde $A_e$ es el área efectiva de la antena receptora.

$A_e = \frac{\lambda^2}{4.\pi}$

Reemplazando una expresión en la otra se llega a:

$P_r = \frac{P_t}{4.\pi.d^2} . A_e = \frac{P_t}{4.\pi.d^2} . \frac{\lambda^2}{4.\pi}$

Finalmente se puede expresar la atenuación del medio (en veces) como:

$\frac{P_t}{P_r} = (\frac{4.\pi.d}{\lambda})^2$

Expresado en DB se lo conoce como la expresión de Friss:

$L_{(dB)} = 20.\log(\frac{4.\pi.d}{\lambda})$

La atenuación del medio dependerá de la longitud de onda $\lambda$ y de la distancia $d$.

La longitud de onda se calcula como:

$\lambda = \frac{c}{F_c} = \frac{3.10^8}{200.10^6} = \frac{3}{2} = 1,5$

La atenuación para los distintos casos será:

$L_{(dB)}|_{5Km} = 20.\log(\frac{4.\pi.5.10^3}{1,5}) = 92,44 dB$

$L_{(dB)}|_{10Km} = 20.\log(\frac{4.\pi.10.10^3}{1,5}) = 98,46 dB$

$L_{(dB)}|_{20Km} = 20.\log(\frac{4.\pi.20.10^3}{1,5}) = 104,48 dB$

La potencia de salida $P_s$ se puede calcular a partir de la potencia de ruido y la relación señal ruido (SNR).

$P_N = 50pW$

$P_{N(dBm)} = 10.\log(\frac{P_N}{1mW}) = 10.\log(\frac{50pW}{1mW}) = -73dBm$

$SNR_{(Veces)} = \frac{P_s}{P_N}$

$SNR_{(dBm)} = P_{s(dBm)}-P_{N(dBm)}$

Finalmente, se puede calcular la potencia de salida como:

$P_{s(dBm)} = SNR_{(dBm)}+P_{N(dBm)} = 33dB + (-73dBm) = -40dBm$

Conociendo la ganancia de los transmisores ($G_A/G_S$), la atenuación del medio ($L_{(dB)}$) y la potencia de salida ($P_s$), se puede calcular la potencia de transmisión como:

$P_{t(dBm)} = P_{s(dBm)} + L_{(dB)} - (G_A+G_S)$

$P_{t(dBm)}|_{5Km} = -40dBm + 92,44 dB - 4,3 dB = 48,14 dBm$

$P_{t(dBm)}|_{10Km} = -40dBm + 98,46 dB - 4,3 dB = 54,16 dBm$

$P_{t(dBm)}|_{20Km} = -40dBm + 104,48 dB - 4,3 dB = 60,18 dBm$

Para convertirlo a potencia en $[W]$ utilizamos:

$P_{(W)} = 1mW . 10^{\frac{P_{(dBm)}}{10}}$

$P_{t(W)}|_{5Km} = 65,16W$

$P_{t(W)}|_{10Km} = 413W$

$P_{t(W)}|_{20Km} = 1042,31W$

Para convertirlo a potencia en $[dBW]$ utilizamos:

$P_{(dBW)} = 10 . \log(\frac{P_{(W)}}{1W})$

$P_{t(dBW)}|_{5Km} = 18,14 dBW$

$P_{t(dBW)}|_{10Km} = 24,16 dBW$

$P_{t(dBW)}|_{20Km} = 30,18 dBW$

## b)

Para el caso en el que el medio es un cable Coaxil, la atenuación varía linealmente con la distancia, siendo esta para cada caso:

$L_{(dB)}|_{5Km} = 5\frac{dB}{Km} . 5Km = 25dB$

$L_{(dB)}|_{10Km} = 5\frac{dB}{Km} . 10Km = 50dB$

$L_{(dB)}|_{20Km} = 5\frac{dB}{Km} . 20Km = 100dB$

Siendo este el único factor que cambia, se calcula la potencia transmitida como:

$P_{t(dBm)} = P_{s(dBm)} + L_{(dB)} - (G_A+G_S)$

$P_{t(dBm)}|_{5Km} = -40dBm + 25 dB - 4,3 dB = -19,3 dBm$

$P_{t(dBm)}|_{10Km} = -40dBm + 50 dB - 4,3 dB = 5,7 dBm$

$P_{t(dBm)}|_{20Km} = -40dBm + 100 dB - 4,3 dB = 55,7 dBm$

Para convertirlo a potencia en $[W]$ utilizamos:

$P_{(W)} = 1mW . 10^{\frac{P_{(dBm)}}{10}}$

$P_{t(W)}|_{5Km} = 11,75 {\mu}W$

$P_{t(W)}|_{10Km} = 3,71 mW$

$P_{t(W)}|_{20Km} = 371,53 W$

Para convertirlo a potencia en $[dBW]$ utilizamos:

$P_{(dBW)} = 10 . \log(\frac{P_{(W)}}{1W})$

$P_{t(dBW)}|_{5Km} =  -49,3dBW$

$P_{t(dBW)}|_{10Km} =  -24,3dBW$

$P_{t(dBW)}|_{20Km} =  25,7dBW$

Para calcular la transmisión de potencia en $[dBmV]$ es necesario calcular el valor de tensión en la entrada para cada caso:

$V_{ief} = \sqrt{P_t.Z_0}$

$V_{ief}|_{5Km} = \sqrt{P_{t(W)}|_{5Km}.Z_0} = \sqrt{11,75.10^{-6}W.75\Omega} = 0,029V$

$V_{ief}|_{10Km} = \sqrt{P_{t(W)}|_{10Km}.Z_0} = \sqrt{3,71.10^{-3}W.75\Omega} = 0,52V$

$V_{ief}|_{20Km} = \sqrt{P_{t(W)}|_{20Km}.Z_0} = \sqrt{371W.75\Omega} = 166,8V$

Ahora se puede calcular la transmisión de potencia como:

$P_{t(dBmV)} = 20.\log(\frac{V_{ief}}{1mV})$

$P_{t(dBmV)}|_{5Km} = 20.\log(\frac{V_{ief}|_{5Km}}{1mV}) = 20.\log(\frac{0,029V}{1mV}) = 29,24 dBmV$

$P_{t(dBmV)}|_{10Km} = 20.\log(\frac{V_{ief}|_{10Km}}{1mV}) = 20.\log(\frac{0,52V}{1mV}) = 54,32 dBmV$

$P_{t(dBmV)}|_{20Km} = 20.\log(\frac{V_{ief}|_{20Km}}{1mV}) = 20.\log(\frac{166,8V}{1mV}) = 104,44 dBmV$
