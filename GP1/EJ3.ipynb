## a)

Se puede expresar la densidad de potencia como:

$\displaystyle S = \frac{P_t}{4.\pi.d^2}$

Y la potencia recibida como:

$\displaystyle P_r = S . A_e$

Donde $A_e$ es el área efectiva de la antena receptora.

$\displaystyle A_e = \frac{\lambda^2}{4.\pi}$

Reemplazando una expresión en la otra se llega a:

$\displaystyle P_r = \frac{P_t}{4.\pi.d^2} . A_e = \frac{P_t}{4.\pi.d^2} . \frac{\lambda^2}{4.\pi}$

Finalmente se puede expresar la atenuación del medio (en veces) como:

$\displaystyle \frac{P_t}{P_r} = \left(\frac{4.\pi.d}{\lambda}\right)^2$

Expresado en DB se lo conoce como la expresión de Friss:

$\displaystyle L_{(dB)} = 20.\log\left(\frac{4.\pi.d}{\lambda}\right)$

La atenuación del medio dependerá de la longitud de onda $\lambda$ y de la distancia $d$.

La longitud de onda se calcula como:

$\displaystyle \lambda = \frac{c}{F_c} = \frac{3.10^8}{200.10^6} = \frac{3}{2} = 1,5$

La atenuación para los distintos casos será:

$\displaystyle L_{(dB)}|_{5Km} = 20.\log\left(\frac{4.\pi.5.10^3}{1,5}\right) = 92,44 dB$

$\displaystyle L_{(dB)}|_{10Km} = 20.\log\left(\frac{4.\pi.10.10^3}{1,5}\right) = 98,46 dB$

$\displaystyle L_{(dB)}|_{20Km} = 20.\log\left(\frac{4.\pi.20.10^3}{1,5}\right) = 104,48 dB$

La potencia de salida $P_s$ se puede calcular a partir de la potencia de ruido y la relación señal ruido (SNR).

$\displaystyle P_N = 50pW$

$\displaystyle P_{N(dBm)} = 10.\log\left(\frac{P_N}{1mW}\right) = 10.\log\left(\frac{50pW}{1mW}\right) = -73dBm$

$\displaystyle SNR_{(Veces)} = \frac{P_s}{P_N}$

$\displaystyle SNR_{(dBm)} = P_{s(dBm)}-P_{N(dBm)}$

Finalmente, se puede calcular la potencia de salida como:

$\displaystyle P_{s(dBm)} = SNR_{(dBm)}+P_{N(dBm)} = 33dB + (-73dBm) = -40dBm$

Conociendo la ganancia de los transmisores ($G_A/G_S$), la atenuación del medio ($L_{(dB)}$) y la potencia de salida ($P_s$), se puede calcular la potencia de transmisión como:

$\displaystyle P_{t(dBm)} = P_{s(dBm)} + L_{(dB)} - (G_A+G_S)$

$\displaystyle P_{t(dBm)}|_{5Km} = -40dBm + 92,44 dB - 4,3 dB = 48,14 dBm$

$\displaystyle P_{t(dBm)}|_{10Km} = -40dBm + 98,46 dB - 4,3 dB = 54,16 dBm$

$\displaystyle P_{t(dBm)}|_{20Km} = -40dBm + 104,48 dB - 4,3 dB = 60,18 dBm$

Para convertirlo a potencia en $[W]$ utilizamos:

$\displaystyle P_{(W)} = 1mW . 10^{\frac{P_{(dBm)}}{10}}$

$\displaystyle P_{t(W)}|_{5Km} = 65,16W$

$\displaystyle P_{t(W)}|_{10Km} = 413W$

$\displaystyle P_{t(W)}|_{20Km} = 1042,31W$

Para convertirlo a potencia en $[dBW]$ utilizamos:

$\displaystyle P_{(dBW)} = 10 . \log\left(\frac{P_{(W)}}{1W}\right)$

$\displaystyle P_{t(dBW)}|_{5Km} = 18,14 dBW$

$\displaystyle P_{t(dBW)}|_{10Km} = 24,16 dBW$

$\displaystyle P_{t(dBW)}|_{20Km} = 30,18 dBW$

## b)

Para el caso en el que el medio es un cable coaxil, la atenuación varía linealmente con la distancia, siendo esta para cada caso:

$\displaystyle L_{(dB)}|_{5Km} = 5\frac{dB}{Km} . 5Km = 25dB$

$\displaystyle L_{(dB)}|_{10Km} = 5\frac{dB}{Km} . 10Km = 50dB$

$\displaystyle L_{(dB)}|_{20Km} = 5\frac{dB}{Km} . 20Km = 100dB$

Considerando la atenuación que produce el coaxil y el hecho de que ya no están presente la ganancia de antenas, se calcula la potencia transmitida como:

$\displaystyle P_{t(dBm)} = P_{s(dBm)} + L_{(dB)}$

$\displaystyle P_{t(dBm)}|_{5Km} = -40dBm + 25 dB = -15 dBm$

$\displaystyle P_{t(dBm)}|_{10Km} = -40dBm + 50 dB = 10 dBm$

$\displaystyle P_{t(dBm)}|_{20Km} = -40dBm + 100 dB = 60 dBm$

Para convertirlo a potencia en $[W]$ utilizamos:

$\displaystyle P_{(W)} = 1mW . 10^{\frac{P_{(dBm)}}{10}}$

$\displaystyle P_{t(W)}|_{5Km} = 31,62 {\mu}W$

$\displaystyle P_{t(W)}|_{10Km} = 10 mW$

$\displaystyle P_{t(W)}|_{20Km} = 1000 W$

Para convertirlo a potencia en $[dBW]$ utilizamos:

$\displaystyle P_{(dBW)} = 10 . \log\left(\frac{P_{(W)}}{1W}\right)$

$\displaystyle P_{t(dBW)}|_{5Km} = -45 dBW$

$\displaystyle P_{t(dBW)}|_{10Km} = -20 dBW$

$\displaystyle P_{t(dBW)}|_{20Km} = 30 dBW$

Para calcular la transmisión de potencia en $[dBmV]$ es necesario calcular el valor de tensión en la entrada para cada caso:

$\displaystyle V_{ief} = \sqrt{P_t.Z_0}$

$\displaystyle V_{ief}|_{5Km} = \sqrt{P_{t(W)}|_{5Km}.Z_0} = \sqrt{31,62.10^{-6}W.75\Omega} = 0,0486V$

$\displaystyle V_{ief}|_{10Km} = \sqrt{P_{t(W)}|_{10Km}.Z_0} = \sqrt{10.10^{-3}W.75\Omega} = 0,866V$

$\displaystyle V_{ief}|_{20Km} = \sqrt{P_{t(W)}|_{20Km}.Z_0} = \sqrt{1000W.75\Omega} = 273,86V$

Ahora se puede calcular la transmisión de potencia como:

$\displaystyle P_{t(dBmV)} = 20.\log\left(\frac{V_{ief}}{1mV}\right)$

$\displaystyle P_{t(dBmV)}|_{5Km} = 20.\log\left(\frac{V_{ief}|_{5Km}}{1mV}\right) = 20.\log\left(\frac{0,0486V}{1mV}\right) = 33,73 dBmV$

$\displaystyle P_{t(dBmV)}|_{10Km} = 20.\log\left(\frac{V_{ief}|_{10Km}}{1mV}\right) = 20.\log\left(\frac{0,866V}{1mV}\right) = 58,75 dBmV$

$\displaystyle P_{t(dBmV)}|_{20Km} = 20.\log\left(\frac{V_{ief}|_{20Km}}{1mV}\right) = 20.\log\left(\frac{273,86V}{1mV}\right) = 108,75 dBmV$
